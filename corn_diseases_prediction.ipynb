```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import cv2
from keras.models import load_model
import numpy as np
from tensorflow.keras.utils import  img_to_array
```

```python
print(tf.__version__)
```

```python
model = tf.keras.models.load_model('model-008.model')
```

# Prediction en temps reel

```python
cap = cv2.VideoCapture(0)
def preprocess_image(image):
    # redimensionner l'image en 224x224 pixels
    resized_image = cv2.resize(image, (260, 260))
    # convertir les pixels en valeurs flottantes et normaliser les pixels en les divisant par 255
    preprocessed_image = np.array(resized_image, dtype=np.float32) / 255.0
    # ajouter une dimension au début du tableau numpy pour créer un lot de données pour le modèle
    preprocessed_image = np.expand_dims(preprocessed_image, axis=0)
    # retourner l'image prétraitée
    return preprocessed_image

while True:
    # capturez une image à partir de la vidéo
    ret, frame = cap.read()
    if not ret:
        print("Erreur de capture d'image")
        break
    # prétraitez l'image
    preprocessed_frame = preprocess_image(frame)
    # faites une prédiction à partir du modèle
    prediction = model.predict(preprocessed_frame)
    # obtenir l'étiquette de la classe prédite
    predicted_class = np.argmax(prediction, axis=1)
    # afficher l'étiquette de la classe prédite sur l'image
    cv2.putText(frame, str(predicted_class), (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    # afficher l'image
    cv2.imshow('Plant Disease Classification', frame)
    # attendez que la touche 'q' soit pressée pour quitter
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

# Prediction avec une image

```python


# Définir les classes de maladies de maïs
classes = {0: "Healthy", 1: "Common rust", 2: "Northern leaf blight", 3: "Gray leaf spot"}

# Charger une image avec OpenCV
img_path = "/home/morningstar/Desktop/personnal_project/test/GrayLeafSpot1.JPG"
img = cv2.imread(img_path)

# Redimensionner l'image pour la faire correspondre à la taille d'entrée du modèle
img = cv2.resize(img, (224, 224))

# Prétraiter l'image en la normalisant et en la redimensionnant
img = np.array(img) / 255.0
img = np.reshape(img, (1, 224, 224, 3))

# Faire une prédiction sur l'image
pred = model.predict(img)

# Récupérer la classe prédite
predicted_class = np.argmax(pred)

# Afficher la maladie prédite
print("Predicted disease: ", classes[predicted_class])
```
